```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,15,13,18,16]
y1=[24,16,9,4,1]
plt.plot(x,y)
plt.plot(x,y1)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.plot(x,y,y1)
plt.show()

```

```python
import pandas as pd
import numpy as np

data={'ID':['1','2','2','4','5'],#duplicate ID + string type
      'Name':['Amit','Riya','Riya','Kiran',None],#missing value
      'Marks':[80,np.nan,75,75,"absent"],#missing value
      'Grade':['A','B','B',None,'B']}#missing +categorial
df=pd.DataFrame(data)
df['Marks']=pd.to_numeric(df['Marks'],errors='coerce') # Moved this line here
print(df)
print(df.isnull)
print(df.isnull().sum())
print(df.notnull())
print(df.dropna(axis=1))
print(df.dropna())
print(df.dropna(how='all'))#drop rows where all values
print(df.dropna(thresh=2))#drop rows at  least 2 missing value
print(df.dropna(subset=['Name']))#drop rows where missing value
#df=df.dropna(subset=['Name'])#creating copy
#df.dropna(subset=['Name'],inplace=True)
print(df)
print(df['Marks'].fillna(df['Marks'].mean()))
print(df['Marks'].fillna(0))
print(df['Marks'].fillna(df['Marks'].median()))
print(df['Marks'].fillna(df['Marks'].mode()[0]))
print(df.duplicated())
print(df.drop_duplicates())
print(df['ID'].astype(int))
print(df["Marks"])
```
